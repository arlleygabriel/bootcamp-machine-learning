```python
# setando o índice
df = df.set_index('datetime')
```

```python
cond = df.index <= '2011-12-31'
df[cond]
```

```python
cond = df.index > '2011-12-31'
df[cond]
```

```python
df.loc[df.index <= '2011-12-31', ['total_count']].sum()
```

```python
df.loc[df.index > '2011-12-31', ['total_count']].sum()
```

```python
df.groupby(by='season').mean()
```

```python
df.groupby(by='hour').mean()
```

```python
df.groupby(by='hour').mean().sort_values(by='total_count')
```

```python
df.groupby(by='weekday').mean().sort_values(by='total_count')
```
